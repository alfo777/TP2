```python
# con los datos del gridsearch defino el mejor modelo de xgboost que encontre:
mejor_rf = RandomForestRegressor(
    n_estimators=3000,
    max_depth=15,
    min_samples_leaf=13,
    min_samples_split=3,
    n_jobs = -1,
    random_state =5 ,
    max_features = "auto"
)
```

```python
#Entreno el modelo:
mejor_rf.fit(train,y)
```

```python
ids = test['id']
test = test.drop(['id','provincia','ciudad'],axis = 1)
```

```python
# realizo la prediccion:
prediction = mejor_rf.predict(test)
```

```python
# creo el dataframe para la sumision:
dataKaggle = pd.DataFrame(ids ,columns={'id'})
```
