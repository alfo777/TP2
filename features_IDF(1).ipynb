# Agrego features que son el puntaje de TF-IDF para los textos en las columnas  descripcion y titulo:

# Para Descripcion:

```python
# voy a definir dos funciones que me van a ayudar a calcular el puntaje segun el IDF de ciertos terminos de una consulta:

# Esta funcion recibe un dataframe, el nombre de una columna que debe guardar textos y una lista con palabras,
# esta se llamara consulta. Devuelve un dataframe con el IDF de cada palabra de la consulta.
def creatTablaIDF(df,col,consulta):
    textos = df.loc[:,['id',col]]
    textos['aux']=1
    textos['palabras'] = textos[col].str.split('[\W_]+')
    textos['palabras'] = textos['palabras'].fillna('')
    rows = list()

    no_sirven = ['lo','la','los','las','el','en','y',
             'a', 'ante', 'bajo', 'cabe', 'con', 'contra', 'de', 'desde',
             'en', 'entre', 'hacia', 'hasta', 'para', 'por', 'según', 'sin',
             'so', 'sobre', 'tras', 'durante', 'mediante', 'versus','en','de','un','una','','1','2','o',
              'que','p','del','muy','al','3','dos','m2','se']
    for row in textos[['id','palabras','aux']].iterrows():
        r = row[1]
        for word in map(str,r.palabras):
            if word not in no_sirven:
                rows.append((r.id,word,r.aux))
    
    palabras = pd.DataFrame(rows,columns=['id','palabra','aux'])
    palabras_aux = palabras.drop('id',axis=1)
    TF = palabras_aux.groupby('palabra').count().reset_index()
    TF.columns=['palabra','cant']
    df_consulta = pd.DataFrame(consulta,columns=['palabra'])
    IDF = pd.merge(df_consulta,TF,on='palabra',how='inner')
    IDF['IDF'] = np.log10(240001/IDF['cant'])
    IDF = IDF.drop('cant',axis=1)
    
    return IDF

#La siguiente funcion recibe un dataframe de features, el nombre de la columna con textos del mismo y un dataframe...
# que guarda IDFS. calcula el peso TF-IDF de cada id.

def crearFeatureIDF(features,col,IDF):
    
    textos = features.loc[:,['id',col]]
    textos['aux']=1
    textos['palabras'] = textos[col].str.split('[\W_]+')
    textos['palabras'] = textos['palabras'].fillna('')
    rows = list()

    no_sirven = ['lo','la','los','las','el','en','y',
             'a', 'ante', 'bajo', 'cabe', 'con', 'contra', 'de', 'desde',
             'en', 'entre', 'hacia', 'hasta', 'para', 'por', 'según', 'sin',
             'so', 'sobre', 'tras', 'durante', 'mediante', 'versus','en','de','un','una','','1','2','o',
              'que','p','del','muy','al','3','dos','m2','se']
    for row in textos[['id','palabras','aux']].iterrows():
        r = row[1]
        for word in map(str,r.palabras):
            if word not in no_sirven:
                rows.append((r.id,word,r.aux))
    palabras = pd.DataFrame(rows,columns=['id','palabra','aux'])
    TF_IDF = pd.merge(IDF,palabras,on='palabra',how='inner')
    TF_IDF = TF_IDF.groupby(['id','palabra','IDF']).count().reset_index()
    TF_IDF['peso_IDF_'+col] = np.log10(1+np.log10(1+TF_IDF['aux']))*TF_IDF['IDF']
    TF_IDF = TF_IDF.loc[:,['id','peso_IDF_'+col]].groupby('id').sum().reset_index()
    features = pd.merge(features,TF_IDF,on='id',how='left')
    features['peso_IDF_'+col] = features['peso_IDF_'+col].fillna(0)
    
    return features
```

```python
# la consulta es:

consulta = ['baño','cocina','casa','sala','comedor','baños','estacionamiento','patio',
            'jardín','cochera','terraza','departamento','terreno','cochera','vigilancia',
            'jardin','techada','juegos','amplio','amplia','amplios','amplias','condominio',
            'nueva','nuevo','garage','gimnasio','jacuzzi','asador','lavandería','jardines',
            'lavanderia','playa','piscina','edificio']
```

```python
# obtengo el feature:
IDFDesc = creatTablaIDF(propiedadesTrain,'descripcion',consulta)
featuresTrain = crearFeatureIDF(featuresTrain,'descripcion',IDFDesc)
featuresTrain.head()
```

# Ahora voy a buscar features segun la columna titulo:

```python
IDFTil = creatTablaIDF(propiedadesTrain,'titulo',consulta)
featuresTrain = crearFeatureIDF(featuresTrain,'titulo',IDFTil)
featuresTrain.head()
```

```python
featuresTrain =featuresTrain.drop('titulo',axis=1)
featuresTrain =featuresTrain.drop('descripcion',axis=1)
```
